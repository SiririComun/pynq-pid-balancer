# Phase 2: System Verification
**Testing the Threaded Balancer and Custom Overlay**

Now that we have encapsulated the hardware, the math, and the threading logic into our `balancin` library, we can interact with the hardware using a clean, professional API.

The most important feature here is **Concurrency**. Because the PID loop is running in a background thread, this notebook will remain responsive.

```python
from balancin.overlay import BalancinOverlay
from balancin.balancer import Balancer
import time

# 1. Load the hardware (Drivers are mapped automatically now!)
ol = BalancinOverlay()

# 2. Initialize the manager
# We start with the constants from the original hardware notebook
bal = Balancer(ol, kp=3.5, ki=0.0, kd=0.0)

print("System ready.")
```

### Starting the Balancer
When you run the cell below, the PID loop will start. You should hear the motor start to react to the sensor angle. 

**Note:** The cell will finish executing immediately, but the motor will keep running in the background!

```python
bal.target_angle = 20.0 # Set the desired angle
bal.start()
```

### Real-Time Interaction
While the motor is running, run the cell below. It will update the $K_p$ constant and the target angle. Observe how the motor's behavior changes instantly without having to restart the loop.

```python
# Change Kp and the target angle on the fly
print("Increasing Kp to 5.0 and changing target to 15 degrees...")
bal.set_params(kp=5.0, target=15.0)

# Wait a bit to observe
time.sleep(3)

print("Returning to original settings...")
bal.set_params(kp=3.5, target=20.0)
```

### Safe Shutdown
Finally, we stop the thread. Our `balancer.stop()` method is designed to join the thread and immediately send a stop signal to the motor.

```python
bal.stop()
```
